### Derivation of SVM Loss w.r.t to W
We have: $$L_i = \sum_{j \neq y_i} max(0, w_j^T x_i - w_{y_i}^T x_i + 1) \quad \text{where $y_i$ is the label of $x_i$}$$

Then by definition:
$$\frac{\partial L_i}{\partial W} = \begin{bmatrix}\frac{\partial L_i}{\partial w_{11}} & \cdots & \frac{\partial L_i}{w_{C1}} \\ \vdots & \cdots & \vdots \\ \frac{\partial L_i}{\partial w_{1D}} & \cdots & \frac{\partial L_i}{w_{CD}}\end{bmatrix} $$

Then:
\begin{align}
    \begin{split}
        \frac{\partial L_i}{\partial w_{11}} &= \frac{\partial \text{ } max(0, w_1^T x_i - w_{y_i}^T x_i + 1)}{\partial w_{11}} \\
        &= \frac{\partial \text{ } max(0, w_{11}x_{i1} + \cdots + w_{1D}x_{iD} - w_{y_{i1}}x_{i1} - \cdots - w_{y_{iD}}x_{iD} + 1)}{\partial w_{11}}
    \end{split}
\end{align}
By Hinge Loss, it is only differentiable when $w_1^T x_i - w_{y_1}^T x_i + 1 > 0$. Thus:

\begin{align}
    \begin{split}
        \frac{\partial L_i}{\partial w_{11}} = x_{i1} \quad \text{if} \text{ } \text{ }w_1^T x_i - w_{y_1}^T x_i + 1 > 0 \quad \text{else} \text{ } \text{ } 0
    \end{split}
\end{align}
Generalizing for cases where $j \neq y_i$ 
$$\frac{\partial L_i}{\partial w_j} =  \begin{bmatrix} x_{i1,} \quad \text{if} \text{ } \text{ }s_j - s_{y_i} + 1 > 0 \quad \text{else} \text{ } \text{ } 0 \\ \vdots \\ x_{iD,} \quad \text{if} \text{ } \text{ }s_j - s_{y_i} + 1 > 0 \quad \text{else} \text{ } \text{ } 0 \end{bmatrix} = x_i \quad \text{if} \text{ } \text{ }s_j - s_{y_i} + 1 > 0 \quad \text{else} \text{ } \text{ } 0$$

Now when $j = y_i$, we evaluate similarly:
$$\frac{\partial L_i}{\partial w_{y_{i1}}}  = -\sum_{j \neq y_i} x_{i1}$$
Generalizing for $y_{ik}$:
$$\frac{\partial L_i}{\partial w_{y_{ik}}}  = -\sum_{j \neq y_i} x_{ik}$$
Thus:
$$\frac{\partial L_i}{\partial w_{y_i}} = -\sum_{j \neq y_i} x_i \quad \text{for all $w_j^Tx_i - w_{y_i}^Tx_i + 1> 0$}$$

It is possible that once in a while, a dimension in the gradcheck will not match exactly. We know the SVM Loss function aka the Hinge Loss is not differentiable at -1 in this case (or generally at sharp peaks of any function $f$), thus computing numeric gradient with value of $f(x + h)$ with value of $h$ is not correctly computed, especially when $h$ is large.

#### Vectorized Implementation:
**Loss**:

We have margins of vectorized SVM Loss that looks mostly like this:
$$\text{margins} = \text{max} \text{ } \left(\text{ } 0, \quad \begin{bmatrix} w_1 \cdot x_1 & w_2 \cdot x_1 & \cdots & w_C \cdot x_1 \\ \vdots & & \cdots & \vdots \\ w_1 \cdot x_N & w_2 \cdot x_N & \cdots & w_C \cdot x_N \end{bmatrix} \text{ } - \text{ }\begin{bmatrix} w_{y_1} \cdot x_1 + 1\\ \vdots \\ w_{y_N} \cdot x_N + 1\end{bmatrix} \text{ } \right)$$

The first term can be obtained by $XW$ and the last term is broadcasted for the entire expression to work. This results in $(N,C)$ matrix. Note that in the implementation we also all the entries $\text{max}(0, w_{y_i} \cdot x_i - w_{y_1} \cdot x_i + 1)$.


**Gradient**:
We have the gradient as following: $$\frac{\partial L}{\partial W} = \begin{bmatrix} \frac{\partial L}{\partial w_1} & \frac{\partial L}{\partial w_2} & \cdots & \frac{\partial L}{\partial w_C} \end{bmatrix}$$
And:
$$\frac{\partial L}{\partial w_j} = \sum_{i = 1}^N \frac{\partial L_i}{\partial w_j} \text{ } \text{$\left(\text{i.e } \sum_{i = 1}^N x_i \text{ }\text{for all } w_j \cdot x_i - w_{y_i} \cdot x_i + 1 > 0 \text{ where } j \neq y_i \right)$}  - \sum x_j \text{ }\left(\text{where } x_j \text{ has label } j \text{ for all } w_k \cdot x_j - w_j \cdot x_j + 1 > 0\right)$$

In order to achieve this, we could turn margins matrix into 0-1 matrix $M$ by filtering out all non-zero margins and update all entries $w_{y_i} \cdot x_i$ to be the sum of non-zero entries on the same row. Thus, the final gradient $\nabla _W L = X^T \cdot M$ where $M$ is the new argins that we just update.

### Stochastic Gradient Descent

We now have vectorized and efficient expressions for the loss and the gradient; and our gradient matches the numerical gradient. We are therefore ready to do SGD to minimize the loss.

```python
# In the file linear_classifier.py, implement SGD in the function
# LinearClassifier.train() and then run it with the code below.
from cs682.classifiers import LinearSVM
svm = LinearSVM()
tic = time.time()
loss_hist = svm.train(X_train, y_train, learning_rate=1e-7, reg=2.5e4,
                      num_iters=1500, verbose=True)
toc = time.time()
print('That took %fs' % (toc - tic))
```

```python
# A useful debugging strategy is to plot the loss as a function of
# iteration number:
plt.plot(loss_hist)
plt.xlabel('Iteration number')
plt.ylabel('Loss value')
plt.show()
```

```python
# Write the LinearSVM.predict function and evaluate the performance on both the
# training and validation set
y_train_pred = svm.predict(X_train)
print('training accuracy: %f' % (np.mean(y_train == y_train_pred), ))
y_val_pred = svm.predict(X_val)
print('validation accuracy: %f' % (np.mean(y_val == y_val_pred), ))
```

```python
# Use the validation set to tune hyperparameters (regularization strength and
# learning rate). You should experiment with different ranges for the learning
# rates and regularization strengths; if you are careful you should be able to
# get a classification accuracy of about 0.4 on the validation set.
learning_rates = [1e-7, 5e-5, 1e-6, 1e-5]
regularization_strengths = [2.5e4, 5e4, 2e4, 3e4, 3.5e4, 4e4, 4.5e4, 5e4]

# results is dictionary mapping tuples of the form
# (learning_rate, regularization_strength) to tuples of the form
# (training_accuracy, validation_accuracy). The accuracy is simply the fraction
# of data points that are correctly classified.
results = {}
best_val = -1   # The highest validation accuracy that we have seen so far.
best_svm = None # The LinearSVM object that achieved the highest validation rate.

################################################################################
# TODO:                                                                        #
# Write code that chooses the best hyperparameters by tuning on the validation #
# set. For each combination of hyperparameters, train a linear SVM on the      #
# training set, compute its accuracy on the training and validation sets, and  #
# store these numbers in the results dictionary. In addition, store the best   #
# validation accuracy in best_val and the LinearSVM object that achieves this  #
# accuracy in best_svm.                                                        #
#                                                                              #
# Hint: You should use a small value for num_iters as you develop your         #
# validation code so that the SVMs don't take much time to train; once you are #
# confident that your validation code works, you should rerun the validation   #
# code with a larger value for num_iters.                                      #
################################################################################

# Cross validation on all combinations of hyperparameters i.e (lr, reg)
for lr in learning_rates:
    for reg in regularization_strengths:
        SVM = LinearSVM()
        # Train SVM
        lost_hist = SVM.train(X_train, y_train, lr, reg, num_iters=1500, verbose=True)
        
        # Training accuracy:
        y_train_pred = SVM.predict(X_train)
        training_accuracy = np.mean(y_train == y_train_pred)
        # Validation accuracy:
        y_val_pred = SVM.predict(X_val)
        validation_accuracy = np.mean(y_val == y_val_pred)
        # Record
        results[(lr, reg)] = (training_accuracy, validation_accuracy)
        # Update
        if validation_accuracy > best_val:
            best_val = validation_accuracy
            best_svm = SVM
        
################################################################################
#                              END OF YOUR CODE                                #
################################################################################
    
# Print out results.
for lr, reg in sorted(results):
    train_accuracy, val_accuracy = results[(lr, reg)]
    print('lr %e reg %e train accuracy: %f val accuracy: %f' % (
                lr, reg, train_accuracy, val_accuracy))
    
print('best validation accuracy achieved during cross-validation: %f' % best_val)
```

```python
# Visualize the cross-validation results
import math
x_scatter = [math.log10(x[0]) for x in results]
y_scatter = [math.log10(x[1]) for x in results]

# plot training accuracy
marker_size = 100
colors = [results[x][0] for x in results]
plt.subplot(2, 1, 1)
plt.scatter(x_scatter, y_scatter, marker_size, c=colors)
plt.colorbar()
plt.xlabel('log learning rate')
plt.ylabel('log regularization strength')
plt.title('CIFAR-10 training accuracy')

# plot validation accuracy
colors = [results[x][1] for x in results] # default size of markers is 20
plt.subplot(2, 1, 2)
plt.scatter(x_scatter, y_scatter, marker_size, c=colors)
plt.colorbar()
plt.xlabel('log learning rate')
plt.ylabel('log regularization strength')
plt.title('CIFAR-10 validation accuracy')
plt.show()
```

```python
# Evaluate the best svm on test set
y_test_pred = best_svm.predict(X_test)
test_accuracy = np.mean(y_test == y_test_pred)
print('linear SVM on raw pixels final test set accuracy: %f' % test_accuracy)
```
