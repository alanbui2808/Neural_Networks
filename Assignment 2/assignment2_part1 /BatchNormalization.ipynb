# Batch Normalization
One way to make deep networks easier to train is to use more sophisticated optimization procedures such as SGD+momentum, RMSProp, or Adam. Another strategy is to change the architecture of the network to make it easier to train. 
One idea along these lines is batch normalization which was proposed by [3] in 2015.

The idea is relatively straightforward. Machine learning methods tend to work better when their input data consists of uncorrelated features with zero mean and unit variance. When training a neural network, we can preprocess the data before feeding it to the network to explicitly decorrelate its features; this will ensure that the first layer of the network sees data that follows a nice distribution. However, even if we preprocess the input data, the activations at deeper layers of the network will likely no longer be decorrelated and will no longer have zero mean or unit variance since they are output from earlier layers in the network. Even worse, during the training process the distribution of features at each layer of the network will shift as the weights of each layer are updated.

The authors of [1] hypothesize that the shifting distribution of features inside deep neural networks may make training deep networks more difficult. To overcome this problem, [3] proposes to insert batch normalization layers into the network. At training time, a batch normalization layer uses a minibatch of data to estimate the mean and standard deviation of each feature. These estimated means and standard deviations are then used to center and normalize the features of the minibatch. A running average of these means and standard deviations is kept during training, and at test time these running averages are used to center and normalize features.

It is possible that this normalization strategy could reduce the representational power of the network, since it may sometimes be optimal for certain layers to have features that are not zero-mean or unit variance. To this end, the batch normalization layer includes learnable shift and scale parameters for each feature dimension.

[1] [Sergey Ioffe and Christian Szegedy, "Batch Normalization: Accelerating Deep Network Training by Reducing
Internal Covariate Shift", ICML 2015.](https://arxiv.org/abs/1502.03167)

### Backward Pass using Computational Graph


**Breakdown of the Computation Graph:**

**[0]**. $x \quad :(N, D)$

**[1]**. $\mu = \displaystyle{\frac{1}{N}} \sum_{i = 1}^N x_i \quad : (D,\text{ })$

**[2]**. $a = x - \mu \quad : (N, D)$

**[3]**. $b = (a_2)^2 = (x - \mu)^2 \quad : (N, D)$

**[4]**. $\sigma^2 = \displaystyle{\frac{1}{N}} \sum_{i = 1}b_i = \frac{\sum_{i =1}^N (x_i - \mu)^2}{N} \quad : (D,\text{ })$

**[5]**. $\sigma = \sqrt{\sigma^2 + \epsilon} = \sqrt{\displaystyle{\frac{\sum_{i =1}^N (x_i - \mu)^2}{N} + \epsilon}} \quad : (D,\text{ })$

**[6]**. $c = 1/\sigma \quad : (D,\text{ })$

**[7]**. $\hat{x} = a_1 \cdot c = \displaystyle{\frac{x - \mu}{\sigma}} \quad : (N, D)$

**[8]**. $e = \gamma \cdot \hat{x} \quad : (N, D)$

**[9]**. $y = e + \beta \quad : (N, D)$

**Backpropagation given $\displaystyle{\frac{\partial L}{\partial y}} \text{ } (N, D)$**: 

Notice that in backpropogation for each step, we first consider each variable to be simply $1D$ variable and proceed to derive its derivative using Chain Rule, then using Matrix Calculus to obtain correct dimension w.r.t to the variable (similar to how we do in Fully Connected Layer part).

**[9]**. $\boxed{y = e + \beta} \quad ; y = (N, D), \text{ }e = (N, D), \text{ } \beta = (D, \text{ })$ given $\displaystyle{\frac{\partial L}{\partial y}} = (N, D)$:

We have: $$\frac{\partial y}{\partial \beta} = 1 \quad \text{and} \quad \frac{\partial y}{\partial e} = 1$$

Thus:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial \beta} &= \frac{\partial L}{\partial y} \cdot \frac{\partial y}{\partial \beta}\\
    &= 1\cdot \frac{\partial L}{\partial f} \\
    \therefore \left( \frac{\partial L}{\partial \beta} \right)_j &= \boxed{\sum_{i=1}^{N} \left( \frac{\partial L}{\partial y}\right)_{i,j}} = \text{sum of column $j$} \quad :(D, \text{ })
\end{split}
\end{align}

And:
$$\frac{\partial L}{\partial e} = \boxed{1 \cdot \frac{\partial L}{\partial y}} \quad :(N, D)$$


**[8]**. $\boxed{e = \gamma \cdot \hat{x}} \quad ; \text{ } e = (N, D), \text{ } \hat{x} = (N,D), \text{ } \gamma = (D, \text{ })$ given $\displaystyle{\frac{\partial L}{\partial e}} = (N, D)$:

We have: $$\frac{\partial e}{\partial \gamma} = \hat{x} \quad \text{and} \quad \frac{\partial e}{\partial \hat{x}} = \gamma$$

Thus:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial \gamma} &= \frac{\partial L}{\partial e} \cdot \frac{\partial e}{\partial \gamma} \\
    &= \frac{\partial L}{\partial e} \odot \hat{x} \quad :(N, D) \\
    \therefore \left( \frac{\partial L}{\partial \gamma} \right)_j &= \boxed{\sum_{i=1}^N \left(\frac{\partial L}{\partial e} \odot \hat{x} \right)_{i, j}} \quad :(D, \text{ })
\end{split}
\end{align}
, where $\odot$ denotes elementwise multiplication.

And:
$$\frac{\partial L}{\partial \hat{x}} = \frac{\partial L}{\partial e} \cdot \frac{\partial e}{\partial \hat{x}} = \boxed{\frac{\partial L}{\partial e} \odot \gamma} \quad :(N, D)$$
Note that $\gamma$ during $\odot$ will get broadcasted.


**[7]**. $\boxed{\hat{x} = a_1 \cdot c = \frac{x - \mu}{\sqrt{\sigma^2 + \epsilon}}} \quad ;\text{ }\hat{x} = (N, D), \text{ } a_1 = (N, D), \text{ } c = (D, \text{ })$ given $\displaystyle{\frac{\partial L}{\partial \hat{x}}} = (N, D)$:

We have: 
$$\frac{\partial \hat{x}}{\partial a_1} = c \quad \text{and} \quad \frac{\partial \hat{x}}{\partial c} = a_1$$

Thus:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial a_1} &= \frac{\partial L}{\partial \hat{x}} \cdot \frac{\partial \hat{x}}{\partial a_1} \\
    &= \boxed{\frac{\partial L}{\partial \hat{x}} \odot c} \quad :(N, D)
\end{split}
\end{align}

And:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial c} &= \frac{\partial L}{\partial \hat{x}} \cdot \frac{\partial \hat{x}}{\partial c} \\
    &= \frac{\partial L}{\partial \hat{x}} \odot a_1  \quad :(N, D)\\
    \therefore \left( \frac{\partial L}{\partial c}\right)_j &= \boxed{\sum_{i=1}^N \left( \frac{\partial L}{\partial \hat{x}} \odot a_1\right)_{i,j}} \quad :(D, \text{ })
\end{split}
\end{align}

**[6]**. $\boxed{c = 1/ \sigma} \quad; \text{ }c = (D, \text{ }), \text{ } \sigma = (D, \text{ })$ given $\frac{\partial L}{\partial c} = (D, \text{ })$:

We have: 
$$ \frac{\partial c}{\partial \sigma} = -\frac{1}{\sigma^2} \quad :(D, \text{ })$$

Thus:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial \sigma} &= \frac{\partial L}{\partial c} \cdot \frac{\partial c}{\partial \sigma} \\
    &= \boxed{\frac{\partial L}{\partial c} \odot -\frac{1}{\sigma^2}} \quad :(D, \text{ }) 
\end{split}
\end{align}

**[5]** $\boxed{\sigma = \sqrt{\sigma^2 + \epsilon}} \quad; \text{ }\sigma = (D, \text{ }), \text{ } \sigma^2 = (D, \text{ })$ given $\displaystyle{\frac{\partial L}{\partial \sigma}} = (D, \text{ })$:

We have:
$$\frac{\partial \sigma}{\partial \sigma^2} = \frac{1}{2\sqrt{\sigma^2 + \epsilon}} \quad :(D, \text{ })$$

Thus:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial \sigma^2} &= \frac{\partial L}{\partial \sigma} \cdot \frac{\partial \sigma}{\partial \sigma^2} \\
    &= \boxed{\frac{\partial L}{\partial \sigma} \odot \frac{1}{2\sqrt{\sigma^2 + \epsilon}}} \quad :(D, \text{ })
\end{split}
\end{align}

**[4]**. $\boxed{\sigma^2 = \frac{1}{N}\sum_{i = 1}^N b_i} \quad;\text{ } b = (N, D), \text{ } \sigma^2 = (D, \text{ })$ given $\displaystyle{\frac{\partial L}{\partial \sigma^2}} = (D, \text{ })$:

We consider $f = \sum_{i=1}^N x_i$ then:
$$ \frac{\partial f}{\partial x_j} = \frac{1}{N}$$

Therefore:
$$\frac{\partial \sigma^2}{\partial b_i} = \frac{1}{N}$$

Thus:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial b} &= \frac{\partial L}{\partial \sigma^2} \cdot \frac{\partial \sigma^2}{\partial b} \\
    &= \boxed{\frac{1}{N}\cdot \begin{bmatrix} 1 & \cdots & 1 \\ \vdots & \ddots & \vdots \\ 1 & \cdots & 1\end{bmatrix} \odot \frac{\partial L}{\partial \sigma^2}} \quad = (N, D) \odot (D, \text{ }) = (N, D)
\end{split}
\end{align}

**[3]** $\boxed{b = (a_2)^2} \quad; \text{ }b = (N, D), \text{ } a_2 = (N, D)$ given $\displaystyle{\frac{\partial L}{\partial b}} = (N, D)$:

We have:
$$\frac{\partial b}{\partial a_2} = 2a_2 \quad :(N, D) $$

Thus:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial a_2} &= \frac{\partial L}{\partial b} \cdot \frac{\partial b}{\partial a_2} \\
    &= \boxed{\frac{\partial L}{\partial b} \odot 2a_2} \quad :(N, D)
\end{split}
\end{align}

**[2]**. $\boxed{a = x - \mu} \quad; \text{ }a = (N, D), \text{ } x = (N, D), \text{ } \mu = (D, \text{ })$ given $\displaystyle{\frac{\partial L}{\partial a_1} = (N, D)}$ and $\displaystyle{\frac{\partial L}{\partial a_2}} = (N, D)$:

We have:
$$\frac{\partial a}{\partial x} = 1 \quad \text{and} \quad \frac{\partial a}{\partial \mu} = -1$$

Thus:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial x} &= \left(\frac{\partial L}{\partial a_1} \cdot \frac{\partial a_1}{\partial x}\right) + \left(\frac{\partial L}{\partial a_2} \cdot \frac{\partial a_2}{\partial x}\right) \\
    &= 1 \cdot \frac{\partial L}{\partial a_1} + 1\cdot \frac{\partial L}{\partial a_2} \\
    &= \boxed{\frac{\partial L}{\partial a_1} + \frac{\partial L}{\partial a_2}} \quad :(N, D)
\end{split}
\end{align}

And:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial \mu} &= \left(\frac{\partial L}{\partial a_1} \cdot \frac{\partial a_1}{\partial \mu}\right) + \left(\frac{\partial L}{\partial a_2} \cdot \frac{\partial a_2}{\partial \mu}\right) \\
    \therefore \left(\frac{\partial L}{\partial \mu}\right)_j &= \sum_{i = 1}^N \left(-\frac{\partial L}{\partial a_1} - \frac{\partial L}{\partial a_2} \right)_{i,j} \\
    &= \boxed{- \sum_{i = 1}^N \left( \frac{\partial L}{\partial a_1} + \frac{\partial L}{\partial a_2}\right)_{i,j}} \quad :(D, \text{ })
\end{split}
\end{align}


**[1]**. $\boxed{\mu = \sum_{i=1}^N x_i} \quad; \text{ }\mu = (D, \text{ }), \text{ } x = (N, D)$ given $\displaystyle{\frac{\partial L}{\partial \mu}} = (D, \text{ })$

This should be similar to **[4]**. Thus:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial x} &= \frac{\partial L}{\partial \mu} \cdot \frac{\partial \mu}{\partial x} \\
    &= \boxed{\frac{1}{N} \mathbb{1} \odot \frac{\partial L}{\partial \mu}} \quad :(N, D)
\end{split}
\end{align}

**[0]**. $\boxed{x} \quad; \text{ }x = (N,D)$ given $\displaystyle{\frac{\partial L}{\partial x_1}} = (N, D)$ from **[1]** and $\displaystyle{\frac{\partial L}{\partial x_2}} = (N, D)$ from **[2]**:  

We have: 
$$\frac{\partial x}{\partial x} = 1 $$

Thus:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial x} &= \left(\frac{\partial L}{\partial x_1} \cdot \frac{\partial x_1}{\partial x_1}\right) + \left( \frac{\partial L}{\partial x_2} \cdot \frac{\partial x_2}{\partial x_2}\right) \\
    &= \boxed{\frac{\partial L}{\partial x_1} + \frac{\partial L}{\partial x_2}} \quad :(N, D)
\end{split}
\end{align}

```python
# Gradient check batchnorm backward pass
np.random.seed(231)
N, D = 4, 5
x = 5 * np.random.randn(N, D) + 12
gamma = np.random.randn(D)
beta = np.random.randn(D)
dout = np.random.randn(N, D)

bn_param = {'mode': 'train'}
fx = lambda x: batchnorm_forward(x, gamma, beta, bn_param)[0]
fg = lambda a: batchnorm_forward(x, a, beta, bn_param)[0]
fb = lambda b: batchnorm_forward(x, gamma, b, bn_param)[0]

dx_num = eval_numerical_gradient_array(fx, x, dout)
da_num = eval_numerical_gradient_array(fg, gamma.copy(), dout)
db_num = eval_numerical_gradient_array(fb, beta.copy(), dout)

_, cache = batchnorm_forward(x, gamma, beta, bn_param)
dx, dgamma, dbeta = batchnorm_backward(dout, cache)
#You should expect to see relative errors between 1e-13 and 1e-8
print('dx error: ', rel_error(dx_num, dx))
print('dgamma error: ', rel_error(da_num, dgamma))
print('dbeta error: ', rel_error(db_num, dbeta))
```

# Batch Normalization: Alternative Backward
In class we talked about two different implementations for the sigmoid backward pass. One strategy is to write out a computation graph composed of simple operations and backprop through all intermediate values. Another strategy is to work out the derivatives on paper. For example, you can derive a very simple formula for the sigmoid function's backward pass by simplifying gradients on paper.

Surprisingly, it turns out that you can do a similar simplification for the batch normalization backward pass too.  
Given a set of inputs $X=\begin{bmatrix}x_1\\x_2\\...\\x_N\end{bmatrix}$, 
we first calculate the mean $\mu=\frac{1}{N}\sum_{k=1}^N x_k$ and variance $v=\frac{1}{N}\sum_{k=1}^N (x_k-\mu)^2.$    
With $\mu$ and $v$ calculated, we can calculate the standard deviation $\sigma=\sqrt{v+\epsilon}$  and normalized data $Y$ with $y_i=\frac{x_i-\mu}{\sigma}.$


The meat of our problem is to get $\frac{\partial L}{\partial X}$ from the upstream gradient $\frac{\partial L}{\partial Y}.$ It might be challenging to directly reason about the gradients over $X$ and $Y$ - try reasoning about it in terms of $x_i$ and $y_i$ first.

You will need to come up with the derivations for $\frac{\partial L}{\partial x_i}$, by relying on the Chain Rule to first calculate the intermediate $\frac{\partial \mu}{\partial x_i}, \frac{\partial v}{\partial x_i}, \frac{\partial \sigma}{\partial x_i},$ then assemble these pieces to calculate $\frac{\partial y_i}{\partial x_i}$. You should make sure each of the intermediary steps are all as simple as possible. 

After doing so, implement the simplified batch normalization backward pass in the function `batchnorm_backward_alt` and compare the two implementations by running the following. Your two implementations should compute nearly identical results, but the alternative implementation should be a bit faster.

### Alternative Backward Derivation
We have by Batch Normalization:
$$ \hat{x}_i = \frac{x_i - \mu}{\sqrt{\sigma^2 + \epsilon}} \quad \text{for $i=1, \cdots, N$ }$$ 
$$ y_i = \gamma \cdot \hat{x}_i + \beta \quad \text{for $i=1, \cdots, N$}$$

Similar to how we do with Computational Graph derivation, we first treat each $x_i$ as 1D variable, once we derive the whole formula, then at the end we generalise it to make sure the dimensions align with the variables.

Since $\mu$, $\sigma^2$ and $\hat{x}_i$ are functions are of $x_i$, thus by Chain Rule we have:
$$\frac{\partial L}{\partial x_i} = \left(\frac{\partial L}{\partial \hat{x}_i} \cdot \frac{\partial \hat{x}_i}{\partial x_i}\right) + \left( \frac{\partial L}{\partial \mu} \cdot \frac{\partial \mu}{\partial x_i}\right) + \left( \frac{\partial L}{\partial \sigma^2} \cdot \frac{\partial \sigma^2}{\partial x_i}\right)$$

**[1]. Computing $\displaystyle{\frac{\partial L}{\partial \hat{x}_i}}$ and $\displaystyle{\frac{\partial \hat{x}_i}{\partial x_i}}$**: 

We have:
$$\boxed{\frac{\partial L}{\partial \hat{x}_i} = \frac{\partial L}{\partial y_i} \cdot \gamma \quad \quad \text{and} \quad \quad \frac{\partial \hat{x}_i}{\partial x_i} = \frac{1}{\sqrt{\sigma^2 + \epsilon}}}$$


**[2]. Computing $\displaystyle{\frac{\partial L}{\partial \mu}}$ and $\displaystyle{\frac{\partial \mu}{\partial x_i}}$**:

**Computing $\displaystyle{\frac{\partial L}{\partial \mu}}$**:

Since $\hat{x}_i$ and $\sigma^2$ are functions are of $\mu$, by Chain Rule we have:
$$\frac{\partial L}{\partial \mu} = \left(\frac{\partial L}{\partial \hat{x}} \cdot \frac{\partial \hat{x}}{\partial \mu}\right) + \left( \frac{\partial L}{\partial \sigma^2} \cdot \frac{\partial \sigma^2}{\partial \mu}\right)$$

Note that we already have $\displaystyle{\frac{\partial L}{\partial \hat{x}}}$ in **[1]**.

First, let's take a look at $\displaystyle{\frac{\partial \hat{x}_i}{\partial \mu}}$:

$$\frac{\partial \hat{x}_i}{\partial \mu} = -\frac{1}{\sqrt{\sigma^2 + \epsilon}}$$ 

Since $\mu$ is contributed by all the $\hat{x}_i$ terms, we have:
$$\boxed{\therefore \frac{\partial L}{\partial \hat{x}} \cdot \frac{\partial \hat{x}}{\partial \mu} = \sum_{i=1}^N \frac{\partial L}{\partial \hat{x}_i} \cdot -\frac{1}{\sqrt{\sigma^2 + \epsilon}}}$$


Second, we have $\sigma^2$ is a function of $\hat{x}$ as following:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial \sigma^2} &= \frac{\partial L}{\partial \hat{x}} \cdot \frac{\partial \hat{x}}{\partial \sigma^2}
\end{split}
\end{align}

Since $\displaystyle{\frac{\partial \hat{x}_i}{\partial \sigma^2} = -\frac{1}{2}(x_i - \mu) \cdot (\sigma^2 + \epsilon)^{-3/2}} $, thus:
$$\frac{\partial \hat{x}}{\partial \sigma^2} = -\frac{1}{2}\sum_{i=1}^N (x_i - \mu) \cdot (\sigma^2 + \epsilon)^{-\frac{3}{2}}  $$

Combining 2 terms, we have:
$$\boxed{\frac{\partial L}{\partial \sigma^2} = -\frac{1}{2}\sum_{i=1}^N \frac{\partial L}{\partial \hat{x}_i} \cdot (x_i - \mu) \cdot (\sigma^2 + \epsilon)^{-3/2}}$$

Third, we have:
\begin{align}
\begin{split}
    \sigma^2 &= \frac{1}{N}\sum_{i=1}^N (x_i - \mu)^2 \\
    \frac{\partial \sigma^2}{\partial \mu} &= \frac{1}{N} \sum_{i=1}^N 2 \cdot (x_i - \mu) \cdot (-1) \\
    &= -2 \cdot \left(\frac{1}{N}\sum_{i=1}^N x_i - \frac{1}{N}\sum_{i=1}^N \mu\right) \\
    &= -2 \cdot \left(\mu - \frac{1}{N} \cdot N\mu\right)\\
    &= 0
\end{split}
\end{align}
$$\therefore \boxed{\frac{\partial \sigma^2}{\partial \mu} = 0 }$$

Putting everything back together, we got:

\begin{align}
\begin{split}
    \frac{\partial L}{\partial \mu} &=  \left(\frac{\partial L}{\partial \hat{x}} \cdot \frac{\partial \hat{x}}{\partial \mu}\right) + \left(\frac{\partial L}{\partial \sigma^2}\cdot \frac{\partial \sigma^2}{\partial \mu} \right) \\
    &= \sum_{i=1}^N \frac{\partial L}{\partial \hat{x}_i} \cdot -\frac{1}{\sqrt{\sigma^2 + \epsilon}} + 0
\end{split}
\end{align}

However, we notice that $\mu$ contributes to all $\hat{x}_i$, thus in generally, we have:
$$\boxed{\frac{\partial L}{\partial \mu} = \sum_{i=1}^N \frac{\partial L}{\partial \hat{x}_i} \cdot -\frac{1}{\sqrt{\sigma^2 + \epsilon}}}$$



**Next compute $\displaystyle{\frac{\partial \mu}{\partial x_i}}$**:

Obviously:
\begin{align}
\begin{split}
    \mu = \frac{1}{N}\sum_{i=1}^N x_i \\
    \therefore \boxed{\frac{\partial \mu}{\partial x_i} = \frac{1}{N}}
\end{split}
\end{align}

**[3]. Computing $\displaystyle{\frac{\partial L}{\partial \sigma^2}}$ and $\displaystyle{\frac{\partial \sigma^2}{\partial x_i}}$**:

While computing $\displaystyle{\frac{\partial L}{\partial \mu}}$, we computed $\displaystyle{\frac{\partial L}{\partial \sigma^2}}$ in **[2]**.

And obviously, we have:
\begin{align}
\begin{split}
    \sigma^2 = \frac{1}{N}\sum_{i=1}^N (x_i - \mu)^2 \\
    \therefore \boxed{\frac{\partial \sigma^2}{\partial x_i} = \frac{2}{N}(x_i - \mu)}
\end{split}
\end{align}

#### Putting [1], [2] and [3] together and simplify:
\begin{align}
\begin{split}
    \boxed{\frac{\partial L}{\partial x_i} = \left(\frac{1}{\sqrt{\sigma^2 + \epsilon}} \cdot \frac{\partial L}{\partial \hat{x}_i}\right) \text{ } + \text{ } \left( -\frac{1}{N\sqrt{\sigma^2 + \epsilon}} \sum_{i=1}^N \frac{\partial L}{\partial \hat{x}_i}\right) \text{ } + \text{ } \left( -\frac{1}{N\sqrt{\sigma^2 + \epsilon}} \cdot \hat{x}_i \sum_{j=1}^N \frac{\partial L}{\partial \hat{x}_j} \cdot \hat{x}_j\right)}
\end{split}
\end{align}

where $\displaystyle{\frac{\partial L}{\partial \hat{x}_i} = \frac{\partial L}{\partial y_i} \cdot \gamma}$.

Along with that:
\begin{align}
\begin{split}
    \frac{\partial L}{\partial \beta} &= \frac{\partial L}{\partial y} \cdot \frac{\partial y}{\partial \beta} \\
    &= \boxed{\sum_{i=1}^N \frac{\partial L}{\partial y_i}}
\end{split}
\end{align}

\begin{align}
\begin{split}
    \frac{\partial L}{\partial \gamma} &= \frac{\partial L}{\partial y} \cdot \frac{\partial y}{\partial \gamma} \\
    &= \boxed{\sum_{i=1}^N \frac{\partial L}{\partial y_i} \cdot \hat{x}_i}
\end{split}
\end{align}



```python
np.random.seed(231)
N, D = 100, 500
x = 5 * np.random.randn(N, D) + 12
gamma = np.random.randn(D)
beta = np.random.randn(D)
dout = np.random.randn(N, D)

bn_param = {'mode': 'train'}
out, cache = batchnorm_forward(x, gamma, beta, bn_param)

t1 = time.time()
dx1, dgamma1, dbeta1 = batchnorm_backward(dout, cache)
t2 = time.time()
dx2, dgamma2, dbeta2 = batchnorm_backward_alt(dout, cache)
t3 = time.time()

print('dx difference: ', rel_error(dx1, dx2))
print('dgamma difference: ', rel_error(dgamma1, dgamma2))
print('dbeta difference: ', rel_error(dbeta1, dbeta2))
print('speedup: %.2fx' % ((t2 - t1) / (t3 - t2)))
```

# Batch Normalization and Initialization
We will now run a small experiment to study the interaction of batch normalization and weight initialization.

The first cell will train 8-layer networks both with and without batch normalization using different scales for weight initialization. The second layer will plot training accuracy, validation set accuracy, and training loss as a function of the weight initialization scale.

```python
np.random.seed(231)
# Try training a very deep net with batchnorm
hidden_dims = [50, 50, 50, 50, 50, 50, 50]
num_train = 1000
small_data = {
  'X_train': data['X_train'][:num_train],
  'y_train': data['y_train'][:num_train],
  'X_val': data['X_val'],
  'y_val': data['y_val'],
}

bn_solvers_ws = {}
solvers_ws = {}
weight_scales = np.logspace(-4, 0, num=20)
for i, weight_scale in enumerate(weight_scales):
  print('Running weight scale %d / %d' % (i + 1, len(weight_scales)))
  bn_model = FullyConnectedNet(hidden_dims, weight_scale=weight_scale, normalization='batchnorm')
  model = FullyConnectedNet(hidden_dims, weight_scale=weight_scale, normalization=None)

  bn_solver = Solver(bn_model, small_data,
                  num_epochs=10, batch_size=50,
                  update_rule='adam',
                  optim_config={
                    'learning_rate': 1e-3,
                  },
                  verbose=False, print_every=200)
  bn_solver.train()
  bn_solvers_ws[weight_scale] = bn_solver

  solver = Solver(model, small_data,
                  num_epochs=10, batch_size=50,
                  update_rule='adam',
                  optim_config={
                    'learning_rate': 1e-3,
                  },
                  verbose=False, print_every=200)
  solver.train()
  solvers_ws[weight_scale] = solver
```

```python
# Plot results of weight scale experiment
best_train_accs, bn_best_train_accs = [], []
best_val_accs, bn_best_val_accs = [], []
final_train_loss, bn_final_train_loss = [], []

for ws in weight_scales:
  best_train_accs.append(max(solvers_ws[ws].train_acc_history))
  bn_best_train_accs.append(max(bn_solvers_ws[ws].train_acc_history))
  
  best_val_accs.append(max(solvers_ws[ws].val_acc_history))
  bn_best_val_accs.append(max(bn_solvers_ws[ws].val_acc_history))
  
  final_train_loss.append(np.mean(solvers_ws[ws].loss_history[-100:]))
  bn_final_train_loss.append(np.mean(bn_solvers_ws[ws].loss_history[-100:]))
  
plt.subplot(3, 1, 1)
plt.title('Best val accuracy vs weight initialization scale')
plt.xlabel('Weight initialization scale')
plt.ylabel('Best val accuracy')
plt.semilogx(weight_scales, best_val_accs, '-o', label='baseline')
plt.semilogx(weight_scales, bn_best_val_accs, '-o', label='batchnorm')
plt.legend(ncol=2, loc='lower right')

plt.subplot(3, 1, 2)
plt.title('Best train accuracy vs weight initialization scale')
plt.xlabel('Weight initialization scale')
plt.ylabel('Best training accuracy')
plt.semilogx(weight_scales, best_train_accs, '-o', label='baseline')
plt.semilogx(weight_scales, bn_best_train_accs, '-o', label='batchnorm')
plt.legend()

plt.subplot(3, 1, 3)
plt.title('Final training loss vs weight initialization scale')
plt.xlabel('Weight initialization scale')
plt.ylabel('Final training loss')
plt.semilogx(weight_scales, final_train_loss, '-o', label='baseline')
plt.semilogx(weight_scales, bn_final_train_loss, '-o', label='batchnorm')
plt.legend()
plt.gca().set_ylim(1.0, 3.5)

plt.gcf().set_size_inches(15, 15)
plt.show()
```

(1). When the initialization weight are too small, the network experiences vanishing gradient. This causes most of the weights in our network to be not updated. However, the batch normalization is less sensitive to small weight initialization.

(2). When the network experiences large weight initialization and thus exploding gradient, the baseline model loss function also explodes.
```python
def run_batchsize_experiments(normalization_mode):
    np.random.seed(231)
    # Try training a very deep net with batchnorm
    hidden_dims = [100, 100, 100, 100, 100]
    num_train = 1000
    small_data = {
      'X_train': data['X_train'][:num_train],
      'y_train': data['y_train'][:num_train],
      'X_val': data['X_val'],
      'y_val': data['y_val'],
    }
    n_epochs=10
    weight_scale = 2e-2
    batch_sizes = [5,10,50]
    lr = 10**(-3.5)
    solver_bsize = batch_sizes[0]

    print('No normalization: batch size = ',solver_bsize)
    model = FullyConnectedNet(hidden_dims, weight_scale=weight_scale, normalization=None)
    solver = Solver(model, small_data,
                    num_epochs=n_epochs, batch_size=solver_bsize,
                    update_rule='adam',
                    optim_config={
                      'learning_rate': lr,
                    },
                    verbose=False)
    solver.train()
    
    bn_solvers = []
    for i in range(len(batch_sizes)):
        b_size=batch_sizes[i]
        print('Normalization: batch size = ',b_size)
        bn_model = FullyConnectedNet(hidden_dims, weight_scale=weight_scale, normalization=normalization_mode)
        bn_solver = Solver(bn_model, small_data,
                        num_epochs=n_epochs, batch_size=b_size,
                        update_rule='adam',
                        optim_config={
                          'learning_rate': lr,
                        },
                        verbose=False)
        bn_solver.train()
        bn_solvers.append(bn_solver)
        
    return bn_solvers, solver, batch_sizes

batch_sizes = [5,10,50]
bn_solvers_bsize, solver_bsize, batch_sizes = run_batchsize_experiments('batchnorm')
```

# Layer Normalization
Batch normalization has proved to be effective in making networks easier to train, but the dependency on batch size makes it less useful in complex networks which have a cap on the input batch size due to hardware limitations. 

Several alternatives to batch normalization have been proposed to mitigate this problem; one such technique is Layer Normalization [2]. Instead of normalizing over the batch, we normalize over the features. In other words, when using Layer Normalization, each feature vector corresponding to a single datapoint is normalized based on the sum of all terms within that feature vector.

[2] [Ba, Jimmy Lei, Jamie Ryan Kiros, and Geoffrey E. Hinton. "Layer Normalization." stat 1050 (2016): 21.](https://arxiv.org/pdf/1607.06450.pdf)
